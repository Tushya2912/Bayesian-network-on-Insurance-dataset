Data Preprocessing (handling missing values, removing duplicates)

```python
for column in data.columns:
    if data[column].dtype == 'object':  # Categorical column
        data[column].fillna(data[column].mode()[0], inplace=True)
    else:  # Numerical column
        data[column].fillna(data[column].median(), inplace=True)

# Remove duplicates
data.drop_duplicates(inplace=True)

# Ensure data consistency (e.g., correcting typos, standardizing categories)
# Example: Standardize the values of a categorical column
data['Gender'] = data['Gender'].str.lower().str.strip()
data['Gender'] = data['Gender'].replace({'female': 'F', 'male': 'M'})

# Verify no missing values remain
print(data.isnull().sum())
```

Transforming Categorical Variables

```python
categorical_columns = data.select_dtypes(include=['object']).columns

# Perform one-hot encoding
data_encoded = pd.get_dummies(data, columns=categorical_columns, drop_first=True)
```

Importing Pgmpy Librarires and performing	Structure Learning

```python
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import HillClimbSearch, BicScore

# Perform structure learning using Hill Climb Search and BIC scoring
hc = HillClimbSearch(data)
best_model = hc.estimate(scoring_method=BicScore(data))

# Print the discovered structure
print("Discovered structure:")
print(best_model.edges())
```

Perfrom	Parameter Learning

```python
from pgmpy.estimators import MaximumLikelihoodEstimator

# Create the Bayesian Network with the discovered structure
model = BayesianNetwork(best_model.edges())

# Perform parameter learning using Maximum Likelihood Estimation
model.fit(data, estimator=MaximumLikelihoodEstimator)

# Print the learned CPDs (Conditional Probability Distributions)
print("Learned CPDs:")
for cpd in model.get_cpds():
    print(cpd)
```

Function for Inference

```python
from pgmpy.inference import VariableElimination

# Perform inference using Variable Elimination
inference = VariableElimination(model)
```
